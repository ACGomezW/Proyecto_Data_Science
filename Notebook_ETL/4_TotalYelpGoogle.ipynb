```python
google = google.rename(columns={'gmap_id':'business_id'})
google = google.rename(columns={'avg_rating':'stars_business'})
google = google.rename(columns={'category':'categories'})
google = google.rename(columns={'name_x':'business_name'})
google = google.rename(columns={'num_of_reviews':'review_count'})
google = google.rename(columns={'time':'review_date'})
google = google.rename(columns={'rating':'stars_by_user'})
google = google.rename(columns={'text':'review_text'})
google = google.rename(columns={'Estado':'state'})
google = google.rename(columns={'MISC':'attributes'})

```

```python
import datetime

google['review_date'] = google['review_date'] / 1000  #

google['review_date'] = google['review_date'].apply(lambda timestamp: datetime.datetime.utcfromtimestamp(timestamp))
```

```python
google['review_date'] = pd.to_datetime(google['review_date'])
google['review_date'] = google['review_date'].dt.date
```

```python
google['business_id'] = google['business_id'].astype(str)
```

```python
google['business_name'] = google['business_name'].astype(str)
```

```python
google['user_id'] = google['user_id'].astype(str)
```

```python
google['stars_by_user'].unique()
```

```python
google['stars_by_user'] = google['stars_by_user'].astype(str)
```

```python
google['stars_business'].unique()
```

```python
google['stars_business'] = google['stars_business'].astype(str)
```

```python
google['categories'].unique()
```

```python
google['categories'] = google['categories'].astype(str)
google['categories'] = google['categories'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

```python
google['categories'].nunique()
```

```python
df_cat=pd.read_csv('/content/drive/My Drive/Proyecto Final Henry/DatasetsCreados/contractor_categories.txt')
```

```python
google['categories'].head(10)
```

```python
for categoria in df_cat['Air conditioning contractor']:
    google['categories'] = google['categories'].str.replace(categoria, categoria)
```

```python
google['attributes'] = google['attributes'].astype(str)
```

```python
google['attributes'].unique()
```

```python
import ast

def safe_eval(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return {}

google['attributes'] = google['attributes'].apply(safe_eval)
google['attribute_values'] = google['attributes'].apply(lambda x: list(x.values()))
```

```python
google['attribute_values'].head(1)
```

```python
print(type(google['attribute_values'].iloc[0]))
```

```python
google['attribute_values'] = google['attribute_values'].astype(str)
```

```python
google['attribute_values'] = google['attribute_values'].str.replace(r'\[|\]|,', '', regex=True)
```

```python
eliminar_columnas=['attributes']
google=google.drop(columns=eliminar_columnas)
```

```python
google = google.rename(columns={'attribute_values':'attributes'})
```
