```python
import pandas as pd
import json
```

```python
ruta_archivo = '/content/drive/My Drive/Proyecto Final Henry/DatasetsOriginales/review.json'
max_filas = 500000

json_lista = []

with open(ruta_archivo, 'r') as file:
    for _ in range(max_filas):
        try:
            json_obj = next(file)
            analisis_obj = json.loads(json_obj)
            json_lista.append(analisis_obj)
        except StopIteration:
            break  # Se detiene cuando se llega al final del archivo
```

```python
dataframes=[]
for json_obj in json_lista:
    df = pd.DataFrame([json_obj])
    dataframes.append(df)

review= pd.concat(dataframes, ignore_index=True)
```

```python
# guardo el resultado en un csv
ruta_guardar = '/content/drive/My Drive/Proyecto Final Henry/DatasetsCreados/review.csv'
review.to_csv(ruta_guardar, index=False)  # index=False evita que se guarde el índice del DataFrame en el archivo
```
