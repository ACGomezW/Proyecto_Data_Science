```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2 Análisis Exploratorio Profundo (EDA)

### 1.1 Abrir dataset Kaggle

```python
df_kaggle = pd.read_csv('df_kaggle.csv', encoding='latin1')
```

### 1.2 Comprensión de Variables económicas partiendo de histogramas

```python
# Crear histogramas
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.hist(df_kaggle['total_income'], bins=10, color='blue', edgecolor='black')
plt.title('Histograma de total_income')
plt.xlabel('total_income')
plt.ylabel('Frecuencia')

plt.subplot(2, 2, 2)
plt.hist(df_kaggle['Mean'], bins=10, color='green', edgecolor='black')
plt.title('Histograma de Mean')
plt.xlabel('Mean')
plt.ylabel('Frecuencia')

plt.subplot(2, 2, 3)
plt.hist(df_kaggle['Median'], bins=10, color='red', edgecolor='black')
plt.title('Histograma de Median')
plt.xlabel('Median')
plt.ylabel('Frecuencia')

plt.subplot(2, 2, 4)
plt.hist(df_kaggle['Stdev'], bins=10, color='purple', edgecolor='black')
plt.title('Histograma de Stdev')
plt.xlabel('Stdev')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

### Análisis del histograma de total_income

El histograma de total_income muestra que la mayoría de los hogares estadounidenses tienen un ingreso total entre $40,000 y $100,000 por año. La distribución es relativamente simétrica, con un pico en el rango de $60,000 a $80,000.

### Análisis del histograma de mean

El histograma de mean muestra que la media del ingreso total de un hogar en los Estados Unidos es de $100,000 por año. Esto significa que la mitad de los hogares estadounidenses tienen un ingreso total superior a $100,000 y la otra mitad tiene un ingreso total inferior a $100,000.

### Análisis del histograma de median

El histograma de median muestra que la mediana del ingreso total de un hogar en los Estados Unidos es de $60,000 por año. Esto significa que la mitad de los hogares estadounidenses tienen un ingreso total superior a $60,000 y la otra mitad tiene un ingreso total inferior a $60,000.

### Análisis del histograma de stdev

El histograma de stdev muestra que la desviación estándar del ingreso total de un hogar en los Estados Unidos es de $20,000 por año. Esto significa que, en promedio, el ingreso total de un hogar estadounidense se aleja de la media en $20,000.

### Conclusiones

Los cuatro histogramas muestran que el ingreso total de los hogares estadounidenses es una variable con una distribución relativamente simétrica. 

Adicional que se pueden mencionar:

- El histograma de total_income muestra que hay una cola larga a la derecha, lo que indica que hay un número relativamente pequeño de hogares con ingresos muy altos.
- El histograma de mean es más estrecho que el histograma de total_income, lo que indica que la mayoría de los hogares se agrupan alrededor de la media.
- El histograma de median es más bajo que el histograma de mean, lo que indica que la mediana es un valor más realista para la mayoría de los hogares.
- El histograma de stdev es relativamente alto, lo que indica que hay una gran variabilidad en el ingreso total de los hogares estadounidenses.

### 1.3 Análisis de la distribución geográfica

```python
# Crear histogramas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df_kaggle['Lat'], bins=20, color='blue', edgecolor='black')
plt.title('Histograma de Latitud (Lat)')
plt.xlabel('Latitud')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(df_kaggle['Lon'], bins=20, color='green', edgecolor='black')
plt.title('Histograma de Longitud (Lon)')
plt.xlabel('Longitud')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Estos dos histogramas muestran que la población estadounidense está concentrada en las regiones costeras y en el este del país. Esto se debe a una serie de factores, como el clima, la geografía, la economía y la historia.

*Clima*

El clima es un factor importante que influye en la distribución de la población. Las regiones costeras suelen tener un clima más templado que las regiones interiores, lo que las hace más atractivas para vivir.

*Geografía*

La geografía también es un factor importante que influye en la distribución de la población. Las regiones costeras suelen tener acceso a puertos, ríos y otros recursos naturales, lo que las hace más favorables para el desarrollo económico.

*Economía*

La economía también es un factor importante que influye en la distribución de la población. Las regiones costeras suelen ser más prósperas que las regiones interiores, lo que las hace más atractivas para la inmigración y la inversión.


**Los dos histogramas muestran que la población estadounidense está concentrada en las regiones costeras y en el este del país. Este patrón de distribución se debe a una serie de factores, como el clima, la geografía, la economía y la historia.**

### 1.4 Distribución de los rango_income

```python
# Calcular la cuenta de cada categoría en 'income_range'
income_range_counts = df_kaggle['income_range'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
income_range_counts.plot(kind='bar', color='dodgerblue')
plt.title('Distribución de "income_range"')
plt.xlabel('Categoría "income_range"')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)  # Para mantener las etiquetas en posición horizontal
plt.show()
```

El gráfico muestra que la distribución de los ingresos de los hogares estadounidenses es desigual. La mayoría de los hogares (50%) se encuentran en los rangos 3 y 4, con ingresos entre $40.000 y $100.000 por año. Una pequeña minoría de hogares (10%) se encuentra en los rangos 1 y 2, con ingresos superiores a 
$100.000 por año.

El gráfico también muestra que la desigualdad de ingresos ha aumentado en los últimos años. El rango 1 ha crecido más rápido que los otros rangos, lo que significa que la brecha entre los hogares de altos ingresos y los hogares de bajos ingresos se ha ampliado.

```python
# Puedes acceder a estas columnas de la siguiente manera:
total_income = df_kaggle['total_income']
latitude = df_kaggle['Lat']
longitude = df_kaggle['Lon']

# Crear un diagrama de dispersión
plt.figure(figsize=(10, 8))
plt.scatter(latitude, longitude, c=total_income, cmap='viridis', s=20, alpha=0.5)
plt.colorbar(label='Total Income')
plt.title('Diagrama de Dispersión: Total Income vs. Ubicación Geográfica')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.show()
```

El gráfico muestra que el ingreso total de los hogares estadounidenses varía considerablemente de un estado a otro. Los estados con los ingresos totales más altos son California, Nueva York y Massachusetts. Los estados con los ingresos totales más bajos son Mississippi, Arkansas y West Virginia.
