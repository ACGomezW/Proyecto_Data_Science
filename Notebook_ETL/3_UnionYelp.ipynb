```python
import re
import numpy as np

def parse_attributes(text):
    if pd.notna(text):
        text = text.replace("'", "\"")

        matches = re.findall(r'\"(\w+)\": \"(True)\"', text)

        attribute_dict = {key: value for key, value in matches}
        return attribute_dict
    else:
        return {}

df_business['attributes'] = df_business['attributes'].apply(parse_attributes)

def extract_true_keys(d):
    return [key for key, value in d.items() if value == 'True']

df_business['attributes_true'] = df_business['attributes'].apply(extract_true_keys)
```
