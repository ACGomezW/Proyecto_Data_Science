```python
import pandas as pd
import json
```

```python
ruta_archivo = '/content/drive/My Drive/Proyecto Final Henry/DatasetsOriginales/checkin.json'

with open(ruta_archivo, 'r') as file:
    json_string = file.read()

json_objeto = json_string.split('\n')
json_lista = []

for json_obj in json_objeto:
    try:
        analisis_obj = json.loads(json_obj)
        json_lista.append(analisis_obj)
    except json.JSONDecodeError:
        pass
```

```python
dataframes=[]
for json_obj in json_lista:
    df = pd.DataFrame([json_obj])
    dataframes.append(df)

checkin= pd.concat(dataframes, ignore_index=True)
```

```python
# guardo el resultado en un csv
ruta_guardar = '/content/drive/My Drive/Proyecto Final Henry/DatasetsCreados/checkin.csv'
checkin.to_csv(ruta_guardar, index=False)  # index=False evita que se guarde el índice del DataFrame en el archivo
```
